```python
import pandas as pd

# Convert columns to datetime (coerce errors to NaT)
cols = [
    'scheduled_departure_datetime_local',
    'scheduled_arrival_datetime_local',
    'actual_departure_datetime_local',
    'actual_arrival_datetime_local'
]

for c in cols:
    flight_data[c] = pd.to_datetime(flight_data[c], errors='coerce')

# Now calculate departure time lag in minutes
flight_data['departure_time_lag_min'] = (
    (flight_data['actual_departure_datetime_local'] - flight_data['scheduled_departure_datetime_local'])
    .dt.total_seconds() / 60
)

# Flight time lag in minutes
scheduled_flight_time = flight_data['scheduled_arrival_datetime_local'] - flight_data['scheduled_departure_datetime_local']
actual_flight_time = flight_data['actual_arrival_datetime_local'] - flight_data['actual_departure_datetime_local']

flight_data['flight_time_lag_min'] = (scheduled_flight_time - actual_flight_time).dt.total_seconds() / 60

```

```python
import optuna
import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

# Split training into train/validation for tuning
X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 1.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 1.0),
        'gamma': trial.suggest_float('gamma', 0.0, 5.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'random_state': 42,
        'objective': 'reg:squarederror'
    }

    model = xgb.XGBRegressor(**params)
    model.fit(
        X_train_split, y_train_split,
        eval_set=[(X_val, y_val)],  # stop if no improvement in 20 rounds
        verbose=False  # suppress output for faster tuning
    )

    y_pred = model.predict(X_val)
    r2 = r2_score(y_val, y_pred)
    return r2  # maximize R²

# Run Optuna study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=25)

print("Best hyperparameters:", study.best_params)
print("Best validation R²:", study.best_value)

# Train final model on full training set
final_model = xgb.XGBRegressor(**study.best_params)
final_model.fit(X_train, y_train)

# Evaluate on test set
y_pred_test = final_model.predict(X_test)
r2_test = r2_score(y_test, y_pred_test)

print("Test R²:", r2_test)



```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import r2_score
import numpy as np

# --- Scale inputs to avoid NaN issues ---
from sklearn.preprocessing import StandardScaler


# --- Build the neural network ---
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')  # linear output for regression
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# --- Custom callback to calculate R2 per epoch ---
class R2Callback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        y_pred = self.model.predict(X_test, verbose=0)
        y_pred_clipped = np.clip(y_pred, 0, 100)  # clip to 0-100
        r2 = r2_score(y_test, y_pred_clipped)
        print(f"Epoch {epoch+1}: R2 score = {r2:.4f}")

# --- Train the model ---
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=32,
    verbose=1,
    callbacks=[R2Callback()]
)

# --- Final predictions ---
y_pred_final = np.clip(model.predict(X_test), 0, 100)
print("Sample predictions:", y_pred_final[:10])
```
